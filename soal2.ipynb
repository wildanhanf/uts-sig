```python
rekap = pd.read_csv('hasil-hitung.csv', sep=',')
```

```python
rekap.info()
rekap.head(10)
```

```python
mean_PBB = rekap['PBB'].mean()
std_PBB = rekap['PBB'].std()

# Hitung upper limit dan lower limit
upper_limit_PBB = abs(mean_PBB + (std_PBB * 1)).round(2)
lower_limit_PBB = abs(mean_PBB - (std_PBB * 1)).round(2)

print(upper_limit_PBB, lower_limit_PBB)
```

```python
def calculate_limits(column_data):
    n = 1
    mean = column_data.mean()
    std = column_data.std()
    
    upper_limit = abs(mean + (std * n)).round(2)
    lower_limit = abs(mean - (std * n)).round(2)
    
    return upper_limit, lower_limit
```

```python
def create_choropleth(df, column, colors, title):
    # Gabungkan dua dataframe berdasarkan kolom 'KECAMATAN'
    merged_df = kab_majalengka.merge(df, on='KECAMATAN')

    # Hitung batas atas dan batas bawah menggunakan calculate_limits
    upper_limit, lower_limit = calculate_limits(merged_df[column])

    # Buat kolom baru untuk menentukan warna berdasarkan kriteria
    merged_df['Color'] = pd.cut(merged_df[column], bins=[-float('inf'), lower_limit, upper_limit, float('inf')],
                                labels=['< {}'.format(lower_limit), '{} - {}'.format(lower_limit, upper_limit), '> {}'.format(upper_limit)])

    cmap = ListedColormap(colors)
    
    # Plot peta choropleth
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    merged_df.plot(column='Color', ax=ax, legend=True, cmap=cmap, edgecolor='black', linewidth=0.2)

    for x, y, label in zip(kab_majalengka.geometry.centroid.x, kab_majalengka.geometry.centroid.y, kab_majalengka['KECAMATAN']):
        ax.text(x, y, label, fontsize=8, ha='center')

    # Judul
    plt.title(title, fontsize=16)

    plt.show()
```

```python
rekap['Pemenang'] = rekap[['Nasdem', 'PKB', 'PKS', 'PDIP', 'Golkar', 'Gerindra', 'Demokrat', 'PAN', 'PPP', 'Hanura', 'PBB', 'PKPI']].idxmax(axis=1)

merged_df = kab_majalengka.merge(rekap, on='KECAMATAN')

color_scheme = {
    'Nasdem': 'khaki', 'PKB': 'green', 'PKS': 'yellow', 'PDIP': 'red',
    'Golkar': 'blue', 'Gerindra': 'white', 'Demokrat': 'purple',
    'PAN': 'orange', 'PPP': 'pink', 'Hanura': 'brown',
    'PBB': 'cyan', 'PKPI': 'magenta'
}

merged_df['Color'] = merged_df['Pemenang'].map(color_scheme)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
merged_df.plot(ax=ax, color=merged_df['Color'], edgecolor='black', linewidth=0.5)

for x, y, label in zip(kab_majalengka.geometry.centroid.x, kab_majalengka.geometry.centroid.y, kab_majalengka['KECAMATAN']):
    ax.text(x, y, label, fontsize=8, ha='center')

# Judul
plt.title("Pemenang Paslon di Setiap Kecamatan", fontsize=16)

plt.show()
```
